### Exercise 6: Conditional Statements

Write a program that takes a variable `grade` (an integer from 0 to 100) and prints the letter grade according to the following scale:

| Score | Grade |
|-------|-------|
| 90–100 | A |
| 80–89 | B |
| 70–79 | C |
| 60–69 | D |
| 50–59 | E |
| Below 50 | F |

*Test your code with at least three different values, including an edge case (e.g. 90).*

````python

def check_grade(grade):
    if grade >= 90:
        print("Grade is A")
    if (grade >= 80) and (grade <= 89):
        print("Grade is B")
    if (grade >= 70) and (grade <= 79):
        print("Grade is C")
    if (grade >= 60) and (grade <= 69):
        print("Grade is D")
    if (grade >= 50) and (grade <= 59):
        print("Grade is E")
    if (grade < 50):
        print("Grade is F")

grade = 3
check_grade(grade)
grade = 60
check_grade(grade)
grade = 79
check_grade(grade)
````

### Exercise 8: While Loop

Write a `while` loop that starts with a variable `balance = 1000` and subtracts a monthly expense of 150€. The loop should stop when the balance goes below 200€. Print the balance after each subtraction and report how many months it took.

````python
balance = 1000 
count = 0

while balance >= 200:
    balance = balance - 150
    print("Balance:", balance)
    count = count + 1

print("it took", count,"months")
````

### Exercise 9: Functions

a) Write a function called `celsius_to_fahrenheit` that takes a temperature in Celsius and returns it in Fahrenheit. The formula is: F = C × 9/5 + 32.

b) Write a function called `describe_temperature` that takes a Celsius value and returns a string: `"freezing"` if below 0, `"cold"` if 0–10, `"mild"` if 10–20, and `"warm"` if above 20.

c) Call both functions with at least two test values and print the results.

````python
#a)

def celsius_to_fahrenheit(celsius):
    fahrenheit = celsius * 9/5 + 32
    return fahrenheit

#b)
def describe_temperature(celsius):
    if celsius < 0:
        return "freezing"
    if (celsius >= 0) and (celsius < 10):
        return "cold"
    if (celsius >= 10) and (celsius <= 20):
        return "mild"
    if (celsius > 20):
        return "warm"
    

#c)
fahrenheit = celsius_to_fahrenheit(10)
print("Temperature after conversion:", fahrenheit)

fahrenheit = celsius_to_fahrenheit(23)
print("Temperature after conversion:", fahrenheit)

celsius = describe_temperature(5)
print("The temperature is", celsius)

celsius = describe_temperature(20)
print("The temperature is", celsius)
````

### Exercise 10: Objects, Methods, and Classes 

You are building a simple system for a university administration office. Define a class called `Course` with the following:

a) Attributes: `name` (string), `credits` (integer), `enrolled_students` (integer), and `max_capacity` (integer).

b) A method `is_full()` that returns `True` if `enrolled_students` has reached `max_capacity`.

c) A method `summary()` that returns a string like `"Python Basics (5 cr) — 32/50 students enrolled"`.

d) Create two `Course` instances (one that is full and one that is not) and print the output of both methods for each.

````python
#a)
class Course:
    def __init__(self, name, credits, enrolled_students, max_capacity):
        self.name = name
        self.credits = credits
        self.enrolled_students = enrolled_students
        self.max_capacity = max_capacity

#b)
    def is_full(self):
        if self.enrolled_students >= self.max_capacity:
            return True
        else:
            return False

#c)
    def summary(self):
        return f"{self.name} ({self.credits} cr) - {self.enrolled_students}/{self.max_capacity} students enrolled" 

#d)
course1 = Course("Principles of Advertising", 6, 33, 60)
course2 = Course("Financial Accounting", 4, 23, 50)

print(course1.is_full())
print(course2.is_full())

print(course1.summary())
print(course2.summary())

        
````
